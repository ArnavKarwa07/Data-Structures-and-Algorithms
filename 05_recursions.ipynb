# Recursion

```python
# Fibonacci sequence - time complexity O(2^n)

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))  # Output: 55
```

```python
#Linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    def __repr__(self):
        return f"Node({self.data})"

Head = Node(1)
A = Node(2)
B = Node(3)

Head.next = A
A.next = B
print(Head)          # Output: Node(1)
```

```python
# Reverse print linked list using recursion - time complexity O(n) and space complexity O(n)
def reverse(head):
    if not head:
        return head
    reverse(head.next)
    print(head)

reverse(Head)
```
